```python
import os
pth = os.path.abspath(os.path.join(os.getcwd(), '../..'))
import sys
sys.path.append(pth)


import torch
from src.collab_compete import  model
# input_size = (1,3,4,84,84)
actor_nn = model.Actor(state_size=24, action_size=2, layer_in_out=[256,256,2], seed=0)
x = torch.rand(size=(20, 24))
y = actor_nn(x)
params = dict(actor_nn.named_parameters())
```
